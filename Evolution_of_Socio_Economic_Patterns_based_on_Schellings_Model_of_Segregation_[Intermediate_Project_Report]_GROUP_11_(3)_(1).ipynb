## Motivation

An agent-based multi-period model is a simulation that models complex systems by simulating the actions and interactions of individual agents over time. In the context of population settlement in a city based on their ethnicities and income levels, an agent-based model can help to understand how these factors affect patterns of segregation and the dynamics of urban neighborhoods over time. The various advantages of such a model in the context of analysing the socio-economic patttern of settlement is as follows :

1. Realistic representation of complex systems: An agent-based model can capture the complexity of real-world systems by representing the behavior of individual agents and how they interact with each other.

2. Improved accuracy: By simulating individual agents and their behavior, an agent-based model can provide more accurate predictions of system behavior than other modeling techniques.

3. Insights into emergent behavior: The model can capture emergent behavior that arises from the interactions between individual agents, which can provide insights into how the system as a whole functions.

4. Useful for policy analysis: The model can be used to evaluate the impact of different policy interventions on a system, which can inform decision-making and help identify the most effective interventions.

## Advanced Schelling's City Segregation Model

We use a modified Schelling's city segregation model, which is an agent-based model that simulates the settlement patterns of individuals based on their preferences for living near others who are similar to them. In the original model, agents are represented as colored dots on a grid, with different colors representing different ethnicities or income levels. Agents move based on their preferences for living near others who are similar to them, and neighborhoods become more segregated over time as agents cluster together based on their similarities. However, in our modified Schellings model, the main objective is to maximize the mean "Happy" function, which is a combination of desired economic as well as social criterias. 

#### -> Who are the AGENTS in our Model?

The agents in our model consist of Individual households of different ethnicities and income levels, who change locations in a period of time(equivalent of rounds) to achieve a satisfactory happiness level.


#### -> What are their CHOICES/STRATEGIES/ACTIONS?

We assign a Happiness Index for each household, which will have a social factor (Ex : living with neighbors of similar ethnicities increases Happiness Index) as well as an economic factor (Ex : living in the vicinity of utility hubs and buildings increases Happiness Index) and play rounds, with people moving into different spots in the city focusing on maximizing their Happiness Index up until a satisfactory level.

#### -> WHY Choose THIS Model?

This model is apt for mapping and analyzing the behavior of population settlement in a city based on their ethnicities and income level for several reasons. It takes into account the preferences of individual agents and how those preferences affect settlement patterns over time. By simulating the behavior of individual agents, the model can capture the complex and dynamic nature of urban neighborhoods and how they change over time.

We area also able to calculate the Index of dissimilarity, average Happiness Index and other criterias across different areas and of the city as a whole and analyse it over a series of rounds, which helps us understand how these different indexes behave over a period of time(rounds).
```python
def initialize_places(num_places,grid_size):
    """
    Initializes the location of places (schools/hospitals) on the grid.

    Args:
        grid_size (int): The size of the grid.
        num_places (int): The number of places to create.

    Returns:
        list: A list of places with 'x', 'y', and 'type' keys.
    """
    places = []
    place_type='school'
    for i in range(num_places):
        x, y = np.random.randint(0, grid_size), np.random.randint(0, grid_size)
        if grid[x, y]==0:
          rand_trig=np.random.random()
          if rand_trig<=0.5:
            grid[x, y]=-1
            place_type = 'school'
          else:
            grid[x, y]=-2
            place_type = 'hospital'
          
          places.append({'x': x, 'y': y, 'type': place_type})
    return places
```

```python
import numpy as np

def calculate_utility(agent, places, max_distance=10):
    """
    Calculates the utility of an agent based on the distance from the nearest school and hospital.

    Args:
        agent (dict): A dictionary representing an agent with 'x', 'y', and 'race' keys.
        places (list): A list of dictionaries representing the locations of places with 'x', 'y', 'facility_type', and 'utility_type' keys.
        max_distance (float): The maximum distance beyond which utility does not improve any further.

    Returns:
        float: The utility of the agent.
    """
    # Calculate distances from places
    school_distances = []
    hospital_distances = []
    for facility in places:
        distance = np.sqrt((facility['x'] - agent['x'])**2 + (facility['y'] - agent['y'])**2)
        if facility['type'] == 'school':
            school_distances.append(distance)
        elif facility['type'] == 'hospital':
            hospital_distances.append(distance)

    # Calculate nearest facility distances
    nearest_school_distance = min(school_distances) if school_distances else np.inf
    nearest_hospital_distance = min(hospital_distances) if hospital_distances else np.inf

    # Calculate utility based on distance from nearest facility
    if nearest_school_distance > max_distance:
        school_utility = 0
    else:
        school_utility = 1.0 / nearest_school_distance
        
    if nearest_hospital_distance > max_distance:
        hospital_utility = 0
    else:
        hospital_utility = 1.0 / nearest_hospital_distance
        
    
    
#     utility = (school_utility + hospital_utility) / 2
    utility = max(school_utility,hospital_utility)

    return utility, min(nearest_school_distance, nearest_hospital_distance)
```

```python
def calculate_similarity(agent,agents):
    """
    Calculates the similarity score between an agent and its neighbors based on race.

    Args:
        agent (dict): The agent for which to calculate the similarity score.
        grid (numpy.ndarray): The grid of agents.

    Returns:
        float: The similarity score between the agent and its neighbors based on race.
    """

    same_race_count =0
    denom=0
    # Iterate over the neighboring agents and count the number of agents with each race
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == dy == 0:
                continue  # Skip the current agent
            x, y = agent['x'] + dx, agent['y'] + dy
            if 0 <= x < grid.shape[0] and 0 <= y < grid.shape[1]:
                if grid[x,y]>0: #not empty/place
                    denom+=1
                    if getagentfromgrid(x,y,agents)['race']==agent['race']:
                        same_race_count +=1

    # Calculate the similarity score based on the number of neighbors with the same race as the agent
    if denom==0:
        return 0
    
    similarity_score = same_race_count / denom

    return similarity_score
```

```python
def getagentfromgrid(x,y,agents):
    for agent in agents:
        if agent['x']==x:
            if agent['y']==y:
                return agent
    return None
    
```

```python
def calculate_segregation_factor(similarity_threshold):
    """
    Calculates the segregation factor based on the current state of the grid.

    Args:
        grid (numpy.ndarray): The grid representing the environment.
        similarity_threshold (float): The similarity threshold for agent movement.

    Returns:
        float: The segregation factor as a proportion of agents with neighbors of the same race.
    """
    counter=0
    for agent in agents:
        simba=calculate_similarity(agent,agents)
        if simba >similarity_threshold:
            counter+=1
    segregation_factor = counter / len(agents)
    return segregation_factor
```

```python
def calculate_income_similarity(agent,agents,threshold):
    """
    Calculates the similarity score between an agent and its neighbors based on income.

    Args:
        agent (dict): The agent for which to calculate the similarity score.
        grid (numpy.ndarray): The grid of agents.

    Returns:
        float: The similarity score between the agent and its neighbors based on race.
    """

    sim_income_count =0
    denom=0
    # Iterate over the neighboring agents and count the number of agents with each race
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == dy == 0:
                continue  # Skip the current agent
            x, y = agent['x'] + dx, agent['y'] + dy
            if 0 <= x < grid.shape[0] and 0 <= y < grid.shape[1]:
                if grid[x,y]>0: #not empty/place
                    denom+=1
                    if abs(getagentfromgrid(x,y,agents)['income']-agent['income'])<threshold:
                        sim_income_count +=1

    # Calculate the similarity score based on the number of neighbors with the same race as the agent
    if denom==0:
        return 0
    
    similarity_score = sim_income_count / denom


    return similarity_score
```

```python
def calculate_income_segregation_factor(threshold, similarity_thres):
    """
    Calculates the segregation factor based on the current state of the grid.

    Args:
        grid (numpy.ndarray): The grid representing the environment.
        similarity_threshold (float): The similarity threshold for agent movement.

    Returns:
        float: The segregation factor as a proportion of agents with neighbors of the same race.
    """
    counter=0
    for agent in agents:
        simba=calculate_income_similarity(agent,agents,threshold)
        if simba >similarity_thres:
            counter+=1
    segregation_factor = counter / len(agents)
    return segregation_factor
```
